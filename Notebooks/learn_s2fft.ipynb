```python
# Config for 64bit precision
from jax.config import config
config.update("jax_enable_x64", True)

import s2fft
import healpy as hp

import scatcovjax.Sphere_lib as sphlib
```

```python
nside = 32
L = 3 * nside -1
f = sphlib.make_hpx_planet(nside=nside, planet='venus', interp=True, normalize=True)

hp.mollview(f, cmap='viridis', title='Data map')
```

```python
# Compute harmonic coefficients
flm = s2fft.forward_jax(f, L=L, nside=nside, sampling='healpix')
```

```python
# Map back to pixel-space signal
fback = s2fft.inverse_jax(flm, L=L, nside=nside, sampling='healpix')

hp.mollview(fback, cmap='viridis', title='Data map')
```

```python
hp.mollview(f-fback, cmap='viridis', title='Data map')
```
